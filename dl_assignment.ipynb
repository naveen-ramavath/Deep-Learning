```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# load data
data = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")

X = data.drop("label", axis=1) / 255.0
y = data["label"]

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# learning rates to test
learning_rates = [0.00001, 0.0001, 0.001, 0.01, 0.1]

train_acc_list = []
val_acc_list = []
loss_list = []

# loop for each learning rate
for lr in learning_rates:
    print("\nTraining with LR =", lr)

    # build model (fresh every time)
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    # compile
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # train
    history = model.fit(X_train, y_train,
                        epochs=5,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=0)

    # store results
    train_acc = history.history['accuracy'][-1]
    val_acc = history.history['val_accuracy'][-1]
    loss = history.history['loss'][-1]

    train_acc_list.append(train_acc)
    val_acc_list.append(val_acc)
    loss_list.append(loss)

    print(f"Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Loss: {loss:.4f}")

# plot results
plt.plot(learning_rates, val_acc_list, marker='o')
plt.xlabel("Learning Rate")
plt.ylabel("Validation Accuracy")
plt.title("Learning Rate vs Accuracy")
plt.show()
```

in this we can check that for small learning rate accuracy is low low for normal learning rate like 0.001 it is great but for 0.01 also accuracy is slow

## with all optimization techniques

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta

# load data
data = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")

X = data.drop("label", axis=1) / 255.0
y = data["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# optimizers list
optimizers = {
    "SGD": SGD(learning_rate=0.01),
    "SGD_Momentum": SGD(learning_rate=0.01, momentum=0.9),
    "Nesterov": SGD(learning_rate=0.01, momentum=0.9, nesterov=True),
    "Adagrad": Adagrad(learning_rate=0.01),
    "RMSProp": RMSprop(learning_rate=0.001),
    "Adadelta": Adadelta(),
    "Adam": Adam(learning_rate=0.001)
}

# results
results = {}

for name, opt in optimizers.items():
    print("\nTraining with", name)

    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    model.compile(optimizer=opt,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train, y_train,
                        epochs=5,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=0)

    val_acc = history.history['val_accuracy'][-1]
    results[name] = val_acc

    print(f"{name} Validation Accuracy: {val_acc:.4f}")

print("\nFinal Results:", results)
```

# Regularization

## L2 regularization

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

# load data
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1) / 255.0
y_test = test["label"]

# model with L2
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,), kernel_regularizer=l2(0.001)))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, validation_split=0.2)

model.evaluate(X_test, y_test)
```

from this we can see that the models train accuracy is 87 and test also 87  means model is learning well and there is no overfitting on data means l2 handling weights on data

## Ensemble Methods

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np

# model 1
model1 = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

# model 2
model2 = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])


model1.compile(optimizer=Adam(0.001),
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

model2.compile(optimizer=Adam(0.001),
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

# train
model1.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)
model2.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)

# evaluate individually
acc1 = model1.evaluate(X_test, y_test, verbose=0)[1]
acc2 = model2.evaluate(X_test, y_test, verbose=0)[1]

print("Model 1 Accuracy:", acc1)
print("Model 2 Accuracy:", acc2)

# predictions
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)

# ensemble (average)
final_pred = (pred1 + pred2) / 2

# convert to labels
final_labels = np.argmax(final_pred, axis=1)

# accuracy of ensemble
ensemble_acc = np.mean(final_labels == y_test)

print("Ensemble Accuracy:", ensemble_acc)
```

in this result we can see that model1 accuracy is 86 and model2 accuracy is 8744  but model1+model2 combining accuracy is 8789 means better that model1 and model2 this is ensemble learning

# CNN

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# load data
train = pd.read_csv("/content/sample_data/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/sample_data/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train
model.fit(X_train, y_train, epochs=5, validation_split=0.2)

# test
print("Test Accuracy:", model.evaluate(X_test, y_test)[1])
```

from this wee can see that the cnn given the more training and testing accuracy than mlp because mlp only learning image in the form of numbers not learning exactly image shapes but cnn finding edges shapes so it is learning about image so the training accuracy and testing accuracy is very high and very close so no overfitting in data.

in this no sudden jumps are also happenning and learning is happenning well and loss also smooth and for epooch 1 and epoch 5 there is so diff in output so in this better feature learning is happenning.

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
# load data
train = pd.read_csv("/content/sample_data/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/sample_data/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    Dropout(0.5),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train
model.fit(X_train, y_train, epochs=5, validation_split=0.2)

# test
print("Test Accuracy:", model.evaluate(X_test, y_test)[1])
```
