```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# load data
data = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")

X = data.drop("label", axis=1) / 255.0
y = data["label"]

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# learning rates to test
learning_rates = [0.00001, 0.0001, 0.001, 0.01, 0.1]

train_acc_list = []
val_acc_list = []
loss_list = []

# loop for each learning rate
for lr in learning_rates:
    print("\nTraining with LR =", lr)

    # build model (fresh every time)
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    # compile
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # train
    history = model.fit(X_train, y_train,
                        epochs=5,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=0)

    # store results
    train_acc = history.history['accuracy'][-1]
    val_acc = history.history['val_accuracy'][-1]
    loss = history.history['loss'][-1]

    train_acc_list.append(train_acc)
    val_acc_list.append(val_acc)
    loss_list.append(loss)

    print(f"Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Loss: {loss:.4f}")

# plot results
plt.plot(learning_rates, val_acc_list, marker='o')
plt.xlabel("Learning Rate")
plt.ylabel("Validation Accuracy")
plt.title("Learning Rate vs Accuracy")
plt.show()
```

in this we can check that for small learning rate accuracy is low low for normal learning rate like 0.001 it is great but for 0.01 also accuracy is slow

## with all optimization techniques

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta

# load data
data = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")

X = data.drop("label", axis=1) / 255.0
y = data["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# optimizers list
optimizers = {
    "SGD": SGD(learning_rate=0.01),
    "SGD_Momentum": SGD(learning_rate=0.01, momentum=0.9),
    "Nesterov": SGD(learning_rate=0.01, momentum=0.9, nesterov=True),
    "Adagrad": Adagrad(learning_rate=0.01),
    "RMSProp": RMSprop(learning_rate=0.001),
    "Adadelta": Adadelta(),
    "Adam": Adam(learning_rate=0.001)
}

# results
results = {}

for name, opt in optimizers.items():
    print("\nTraining with", name)

    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    model.compile(optimizer=opt,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train, y_train,
                        epochs=5,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=0)

    val_acc = history.history['val_accuracy'][-1]
    results[name] = val_acc

    print(f"{name} Validation Accuracy: {val_acc:.4f}")

print("\nFinal Results:", results)
```

# Regularization

## L2 regularization

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

# load data
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1) / 255.0
y_test = test["label"]

# model with L2
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,), kernel_regularizer=l2(0.001)))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, validation_split=0.2)

model.evaluate(X_test, y_test)
```

from this we can see that the models train accuracy is 87 and test also 87  means model is learning well and there is no overfitting on data means l2 handling weights on data

## Ensemble Methods

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np

# model 1
model1 = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

# model 2
model2 = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])


model1.compile(optimizer=Adam(0.001),
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

model2.compile(optimizer=Adam(0.001),
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

# train
model1.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)
model2.fit(X_train, y_train, epochs=3, batch_size=32, verbose=0)

# evaluate individually
acc1 = model1.evaluate(X_test, y_test, verbose=0)[1]
acc2 = model2.evaluate(X_test, y_test, verbose=0)[1]

print("Model 1 Accuracy:", acc1)
print("Model 2 Accuracy:", acc2)

# predictions
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)

# ensemble (average)
final_pred = (pred1 + pred2) / 2

# convert to labels
final_labels = np.argmax(final_pred, axis=1)

# accuracy of ensemble
ensemble_acc = np.mean(final_labels == y_test)

print("Ensemble Accuracy:", ensemble_acc)
```

in this result we can see that model1 accuracy is 86 and model2 accuracy is 8744  but model1+model2 combining accuracy is 8789 means better that model1 and model2 this is ensemble learning

# CNN

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# load data
train = pd.read_csv("/content/sample_data/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/sample_data/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train
model.fit(X_train, y_train, epochs=5, validation_split=0.2)

# test
print("Test Accuracy:", model.evaluate(X_test, y_test)[1])
```

from this wee can see that the cnn given the more training and testing accuracy than mlp because mlp only learning image in the form of numbers not learning exactly image shapes but cnn finding edges shapes so it is learning about image so the training accuracy and testing accuracy is very high and very close so no overfitting in data.

in this no sudden jumps are also happenning and learning is happenning well and loss also smooth and for epooch 1 and epoch 5 there is so diff in output so in this better feature learning is happenning.

# Pre trained models

## flow

Pretrained Model → Remove last layer → Add new layer → Train

```python
import pandas as pd
import tensorflow as tf

train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# reduce data (important)
X_train = X_train[:5000]
y_train = y_train[:5000]
X_test = X_test[:1000]
y_test = y_test[:1000]

# resize for pretrained
X_train = tf.image.resize(X_train, (224,224))
X_train = tf.repeat(X_train, 3, axis=-1)

X_test = tf.image.resize(X_test, (224,224))
X_test = tf.repeat(X_test, 3, axis=-1)
```

## lenet

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, Flatten, Dense

model = Sequential([
    Conv2D(6, (5,5), activation='relu', input_shape=(224,224,3)),
    AveragePooling2D((2,2)),
    Conv2D(16, (5,5), activation='relu'),
    AveragePooling2D((2,2)),
    Flatten(),
    Dense(120, activation='relu'),
    Dense(84, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=3)
print("LeNet:", model.evaluate(X_test, y_test)[1])
```

in this training accuracy incresed as 74,84,88  it is very smooth learning acuuracy and loss also from 73 to 30 very smooth decrese but test accuracy is 84 so slight overfitting is there but acceptable ,so this is simple model and learn basic features.

## vggnet

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

base = VGG16(weights='imagenet', include_top=False,input_shape=(224,224,3))

for layer in base.layers:
    layer.trainable = False

x = Flatten()(base.output)
x = Dense(64, activation='relu')(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=base.input, outputs=output)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=3)
print("VGG:", model.evaluate(X_test, y_test)[1])
```

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# reduce size (fast)
X_train = X_train[:5000]
y_train = y_train[:5000]
X_test = X_test[:1000]
y_test = y_test[:1000]

# ----------------------------
# 2. BUILD CNN (Functional API)
# ----------------------------
inputs = Input(shape=(28,28,1))

x = Conv2D(32, (3,3), activation='relu')(inputs)
x = MaxPooling2D((2,2))(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
outputs = Dense(10, activation='softmax')(x)

model = Model(inputs, outputs)

# compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# ----------------------------
# 3. TRAIN
# ----------------------------
model.fit(X_train, y_train, epochs=3, validation_split=0.2)

# ----------------------------
# 4. VISUALIZATION MODEL
# ----------------------------
# take first conv layer
conv_layer = model.layers[1]

vis_model = Model(inputs=model.input, outputs=conv_layer.output)

# ----------------------------
# 5. GET FEATURE MAPS
# ----------------------------
img = X_test[0:1]
feature_maps = vis_model.predict(img)

# ----------------------------
# 6. PLOT
# ----------------------------
plt.figure(figsize=(8,5))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(feature_maps[0,:,:,i], cmap='gray')
    plt.title(f"Filter {i}")
    plt.axis('off')

plt.show()
```

The convolutional layer extracts different features using multiple filters
Some filters highlight edges while others capture shapes and textures
Bright regions in feature maps indicate strong activation
Dark regions indicate less important areas
Early layers detect low-level features like edges and outlines

# Guided backpropagation

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_train = train["label"]

X_test = test.drop("label", axis=1).values.reshape(-1,28,28,1) / 255.0
y_test = test["label"]

# small dataset
X_train = X_train[:5000]
y_train = y_train[:5000]
X_test = X_test[:1000]
y_test = y_test[:1000]

# ----------------------------
# 2. BUILD MODEL
# ----------------------------
inputs = Input(shape=(28,28,1))

x = Conv2D(32, (3,3), activation='relu')(inputs)
x = MaxPooling2D((2,2))(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
outputs = Dense(10, activation='softmax')(x)

model = Model(inputs, outputs)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train
model.fit(X_train, y_train, epochs=3, validation_split=0.2)

# ----------------------------
# 3. GUIDED BACKPROP
# ----------------------------

# pick one image
img = X_test[0:1]
img_tensor = tf.convert_to_tensor(img)

with tf.GradientTape() as tape:
    tape.watch(img_tensor)
    preds = model(img_tensor)

    # get predicted class
    class_idx = tf.argmax(preds[0])
    loss = preds[:, class_idx]

# compute gradients
grads = tape.gradient(loss, img_tensor)

# guided backprop: keep only positive gradients
guided_grads = tf.maximum(grads, 0)

# convert to numpy
guided_grads = guided_grads.numpy()[0]

# ----------------------------
# 4. VISUALIZATION
# ----------------------------
plt.figure(figsize=(8,3))

# original image
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img[0].reshape(28,28), cmap='gray')
plt.axis('off')

# guided backprop image
plt.subplot(1,2,2)
plt.title("Guided Backprop")
plt.imshow(guided_grads.reshape(28,28), cmap='gray')
plt.axis('off')

plt.show()
```

Guided backpropagation highlights important pixels contributing to prediction
Bright regions correspond to edges and object boundaries
Background regions are mostly ignored
The model focuses on shape and structure of the object
Only a subset of pixels influence the final prediction

# Encoder and decoder

auto encoder

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values / 255.0
X_test = test.drop("label", axis=1).values / 255.0

# small dataset (fast)
X_train = X_train[:5000]
X_test = X_test[:1000]

# ----------------------------
# 2. BUILD AUTOENCODER
# ----------------------------

# input layer (784 pixels)
input_img = Input(shape=(784,))

# -------- Encoder --------
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)   # latent space

# -------- Decoder --------
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(784, activation='sigmoid')(decoded)

# full model
autoencoder = Model(input_img, decoded)

# ----------------------------
# 3. COMPILE
# ----------------------------
autoencoder.compile(optimizer='adam',
                    loss='mse')

# ----------------------------
# 4. TRAIN
# ----------------------------
autoencoder.fit(X_train, X_train,
                epochs=5,
                batch_size=32,
                validation_split=0.2)

# ----------------------------
# 5. PREDICT (RECONSTRUCTION)
# ----------------------------
reconstructed = autoencoder.predict(X_test)

# ----------------------------
# 6. VISUALIZATION
# ----------------------------
plt.figure(figsize=(8,4))

for i in range(5):
    # original
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # reconstructed
    plt.subplot(2,5,i+6)
    plt.imshow(reconstructed[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show()
```

The autoencoder successfully reconstructs input images from compressed representations
The reconstructed images preserve overall shape but lose fine details
Blurring occurs due to dimensionality reduction
Training and validation loss decrease steadily, indicating proper learning
The small gap between training and validation loss shows no overfitting

## Undercomplete Autoencoder

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values / 255.0
X_test = test.drop("label", axis=1).values / 255.0

# small dataset
X_train = X_train[:5000]
X_test = X_test[:1000]

# ----------------------------
# 2. BUILD UNDERCOMPLETE AE
# ----------------------------
input_img = Input(shape=(784,))

# Encoder (strong compression)
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(16, activation='relu')(encoded)   # 🔥 small latent space

# Decoder
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(784, activation='sigmoid')(decoded)

# Model
autoencoder = Model(input_img, decoded)

# ----------------------------
# 3. COMPILE
# ----------------------------
autoencoder.compile(optimizer='adam', loss='mse')

# ----------------------------
# 4. TRAIN
# ----------------------------
history = autoencoder.fit(X_train, X_train,
                          epochs=5,
                          batch_size=32,
                          validation_split=0.2)

# ----------------------------
# 5. RECONSTRUCTION
# ----------------------------
reconstructed = autoencoder.predict(X_test)

# ----------------------------
# 6. VISUALIZATION
# ----------------------------
plt.figure(figsize=(8,4))

for i in range(5):
    # original
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # reconstructed
    plt.subplot(2,5,i+6)
    plt.imshow(reconstructed[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show()
```

Undercomplete autoencoder compresses input into a smaller latent space
As latent dimension decreases, reconstruction quality reduces
Important features like shape are preserved, but fine details are lost
Higher compression leads to more information loss
There is a trade-off between compression and reconstruction quality

## Overcomplete Autoencoder

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values / 255.0
X_test = test.drop("label", axis=1).values / 255.0

# small dataset
X_train = X_train[:5000]
X_test = X_test[:1000]

# ----------------------------
# 2. BUILD OVERCOMPLETE AE
# ----------------------------
input_img = Input(shape=(784,))

# Encoder (EXPANSION)
encoded = Dense(512, activation='relu')(input_img)
encoded = Dense(1024, activation='relu')(encoded)   # 🔥 bigger than input

# Decoder
decoded = Dense(512, activation='relu')(encoded)
decoded = Dense(784, activation='sigmoid')(decoded)

# Model
autoencoder = Model(input_img, decoded)

# ----------------------------
# 3. COMPILE
# ----------------------------
autoencoder.compile(optimizer='adam', loss='mse')

# ----------------------------
# 4. TRAIN
# ----------------------------
history = autoencoder.fit(X_train, X_train,
                          epochs=5,
                          batch_size=32,
                          validation_split=0.2)

# ----------------------------
# 5. RECONSTRUCTION
# ----------------------------
reconstructed = autoencoder.predict(X_test)

# ----------------------------
# 6. VISUALIZATION
# ----------------------------
plt.figure(figsize=(8,4))

for i in range(5):
    # original
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # reconstructed
    plt.subplot(2,5,i+6)
    plt.imshow(reconstructed[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show()
```

Overcomplete autoencoder has latent dimension larger than input  that we can see we taken output as 1024,
It produces better reconstruction with less information loss
Images appear sharper compared to undercomplete AE
The model may memorize input instead of learning meaningful features
Overcomplete AE requires regularization to prevent overfitting

## regularized ae

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.regularizers import l2

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values / 255.0
X_test = test.drop("label", axis=1).values / 255.0

# small dataset
X_train = X_train[:5000]
X_test = X_test[:1000]

# ----------------------------
# 2. BUILD REGULARIZED AE
# ----------------------------
input_img = Input(shape=(784,))

# Encoder with regularization
encoded = Dense(512, activation='relu', kernel_regularizer=l2(0.001))(input_img)
encoded = Dropout(0.3)(encoded)
encoded = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(encoded)

# Decoder
decoded = Dense(512, activation='relu')(encoded)
decoded = Dense(784, activation='sigmoid')(decoded)

# Model
autoencoder = Model(input_img, decoded)

# ----------------------------
# 3. COMPILE
# ----------------------------
autoencoder.compile(optimizer='adam', loss='mse')

# ----------------------------
# 4. TRAIN
# ----------------------------
history = autoencoder.fit(X_train, X_train,
                          epochs=5,
                          batch_size=32,
                          validation_split=0.2)

# ----------------------------
# 5. RECONSTRUCTION
# ----------------------------
reconstructed = autoencoder.predict(X_test)

# ----------------------------
# 6. VISUALIZATION
# ----------------------------
plt.figure(figsize=(8,4))

for i in range(5):
    # original
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # reconstructed
    plt.subplot(2,5,i+6)
    plt.imshow(reconstructed[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show()
```

in overcomplete ae the model is memorizing so for that we use this model,Regularization prevents the model from memorizing input data
L2 regularization penalizes large weights and improves generalization
Dropout forces the model to learn robust features
Reconstruction becomes smoother and less overfitted
Slight increase in loss is observed due to constraints

## denoising autoencoder

```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# ----------------------------
# 1. LOAD DATA
# ----------------------------
train = pd.read_csv("/content/fashion-mnist_train.csv (1).zip")
test = pd.read_csv("/content/fashion-mnist_test.csv (1).zip")

X_train = train.drop("label", axis=1).values / 255.0
X_test = test.drop("label", axis=1).values / 255.0

# small dataset
X_train = X_train[:5000]
X_test = X_test[:1000]

# ----------------------------
# 2. ADD NOISE
# ----------------------------
noise_factor = 0.3   # 🔥 change this (0.1, 0.3, 0.5)

X_train_noisy = X_train + noise_factor * np.random.normal(size=X_train.shape)
X_test_noisy = X_test + noise_factor * np.random.normal(size=X_test.shape)

# keep values between 0 and 1
X_train_noisy = np.clip(X_train_noisy, 0., 1.)
X_test_noisy = np.clip(X_test_noisy, 0., 1.)

# ----------------------------
# 3. BUILD AUTOENCODER
# ----------------------------
input_img = Input(shape=(784,))

encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(784, activation='sigmoid')(decoded)

autoencoder = Model(input_img, decoded)

# ----------------------------
# 4. COMPILE
# ----------------------------
autoencoder.compile(optimizer='adam', loss='mse')

# ----------------------------
# 5. TRAIN (IMPORTANT)
# ----------------------------
# input = noisy, output = clean
autoencoder.fit(X_train_noisy, X_train,
                epochs=5,
                batch_size=32,
                validation_split=0.2)

# ----------------------------
# 6. RECONSTRUCTION
# ----------------------------
reconstructed = autoencoder.predict(X_test_noisy)

# ----------------------------
# 7. VISUALIZATION
# ----------------------------
plt.figure(figsize=(10,5))

for i in range(5):
    # noisy
    plt.subplot(3,5,i+1)
    plt.imshow(X_test_noisy[i].reshape(28,28), cmap='gray')
    plt.title("Noisy")
    plt.axis('off')

    # original
    plt.subplot(3,5,i+6)
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # reconstructed
    plt.subplot(3,5,i+11)
    plt.imshow(reconstructed[i].reshape(28,28), cmap='gray')
    plt.title("Reconstructed")
    plt.axis('off')

plt.show()
```

Denoising autoencoder reconstructs clean images from noisy inputs
Model learns robust feature representations
Moderate noise levels can be effectively removed
High noise levels reduce reconstruction quality
The model preserves shape but may lose fine details
